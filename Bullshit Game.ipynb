```python
import random

# Set up the game
deck = []
for suit in ['hearts', 'diamonds', 'clubs', 'spades']:
    for rank in range(1, 14):
        deck.append((rank, suit))
random.shuffle(deck)

players = ['Player A', 'Player B']
hands = {player: [] for player in players}
for i in range(len(deck)):
    player = players[i % len(players)]
    hands[player].append(deck[i])

# Play the game
num_cards = len(deck) // len(players)
current_rank = None
current_player = 0
num_cards_played = 0
while True:
    # Print the current state of the game
    print('Current player:', players[current_player])
    print('Current rank:', current_rank)
    print('Number of cards played:', num_cards_played)
    print('Hands:')
    for player in players:
        print(player, len(hands[player]), hands[player])
    print()

    # Ask the current player to play a card
    while True:
        card_index = input('Enter the index of the card you want to play (0-{}): '.format(len(hands[players[current_player]]) - 1))
        if not card_index.isdigit() or int(card_index) < 0 or int(card_index) >= len(hands[players[current_player]]):
            print('Invalid input. Please enter a number between 0 and {}.'.format(len(hands[players[current_player]]) - 1))
        else:
            card_index = int(card_index)
            card = hands[players[current_player]][card_index]
            if current_rank is None or card[0] == current_rank:
                break
            else:
                print('You cannot play that card. You must play a card of rank {}.'.format(current_rank))

    # Remove the played card from the player's hand
    hands[players[current_player]].pop(card_index)

    # Check if the player is calling bullshit
    if random.random() < 0.1:
        print('{} is calling bullshit!'.format(players[current_player]))
        num_cards_played += 1
        if current_rank is not None:
            for player in players:
                if player != players[current_player]:
                    for card in hands[player]:
                        if card[0] == current_rank:
                            hands[player].remove(card)
                            num_cards_played += 1
        if num_cards_played == num_cards:
            print('Game over!')
            break
        else:
            current_player = (current_player + 1) % len(players)
            current_rank = None
    else:
        print('{} played {}.'.format(players[current_player], card))
        num_cards_played += 1
        if num_cards_played == num_cards:
            print('Game over!')
            break
        else:
            current_player = (current_player + 1) % len(players)
            current_rank = card[0]
```
